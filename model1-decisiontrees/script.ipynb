** Predicting Health From GDP (2015)**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

# Decision Tree Model (in-sample)
y = happiness2015['Health (Life Expectancy)']

predictors = ['Economy (GDP per Capita)']
X = happiness2015[predictors]

# Define model
model = DecisionTreeRegressor()

# Fit model
model.fit(X, y)

print("Making predictions for the following entries:")
print(X.head())
print(y.head())
print("The predictions are")
print(model.predict(X.head()))

# The following measure shows an "in-sample" score (this is bad)
predicted_value = model.predict(X)
print("Mean absolute error:", mean_absolute_error(y, predicted_value))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# Decision Tree Model (in-sample)
y = happiness2015['Health (Life Expectancy)']

predictors = ['Economy (GDP per Capita)']
X = happiness2015[predictors]

# split data into training and validation data, for both predictors and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
# Define model
model2 = DecisionTreeRegressor()
# Fit model
model2.fit(train_X, train_y)


print("Making predictions for the following entries:")
print(val_X.head())
print(val_y.head())
print("The predictions are")
print(model2.predict(val_X.head()))


# get predicted prices on validation data
val_predictions = model2.predict(val_X)
print("Mean absolute error:", mean_absolute_error(val_y, val_predictions))
```

```python
# Underfitting, Overfitting and Model Optimization

from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 10, 20, 25, 30, 35, 36, 37, 38, 40, 50, 500]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %f  \t\t Mean Absolute Error:  %f" %(max_leaf_nodes, my_mae))
```

```python
# Random Forest

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
predictions = forest_model.predict(val_X)

print("Making predictions for the following entries:")
print(val_X.head())
print(val_y.head())
print("The predictions are")
print(model2.predict(val_X.head()))

print("Mean absolute error:", mean_absolute_error(val_y, predictions))
```
